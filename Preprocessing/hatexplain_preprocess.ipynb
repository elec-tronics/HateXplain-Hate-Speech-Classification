```python
import pandas as pd
```

```python
df = pd.read_json('hatexplain.json',orient='index')
df.head()
```

```python
df['label_1'] = None
df['label_2'] = None
df['label_3'] = None
df['vote'] = None
```

```python
df.shape
```

```python
for i in range(df.shape[0]):
    df['label_1'][i] = df['annotators'][i][0]['label']
    df['label_2'][i] = df['annotators'][i][1]['label']
    df['label_3'][i] = df['annotators'][i][2]['label']
```

```python
df.loc[((df['label_1']=='normal')&(df['label_2']=='normal'))|
       ((df['label_2']=='normal')&(df['label_3']=='normal'))|
       ((df['label_1']=='normal')&(df['label_3']=='normal')),'vote'] = 'normal'
df.loc[((df['label_1']=='offensive')&(df['label_2']=='offensive'))|
       ((df['label_2']=='offensive')&(df['label_3']=='offensive'))|
       ((df['label_1']=='offensive')&(df['label_3']=='offensive')),'vote'] = 'offensive'
df.loc[((df['label_1']=='hatespeech')&(df['label_2']=='hatespeech'))|
       ((df['label_2']=='hatespeech')&(df['label_3']=='hatespeech'))|
       ((df['label_1']=='hatespeech')&(df['label_3']=='hatespeech')),'vote'] = 'hatespeech'
```

```python
df_out = df[['post_id','post_tokens','label_1','label_2','label_3','vote']]
df_out
```

```python
df_out.loc[df['vote'].isnull(),'vote'] = 'ambiguous'
```

```python
df_out['vote_ordinal'] = 0

df_out.loc[df['vote']=='normal','vote_ordinal'] = 1
df_out.loc[df['vote']=='offensive','vote_ordinal'] = 2
df_out.loc[df['vote']=='hatespeech','vote_ordinal'] = 3
df_out
```

```python
df_out.to_csv('hatexplain_final.csv',index=False)
```

```python
hate_speech['token_sentence'] = hate_speech['post_tokens'].apply(eval).apply(' '.join) #converting post_tokens to sentence
```

```python
hate_speech['token_sentence']
```

```python
from nltk import word_tokenize
import nltk
#nltk.download('punkt')
```

```python
hate_speech['new_token_sentence'] = hate_speech['token_sentence'].apply(word_tokenize)
```

```python
from nltk.probability import FreqDist
frequency = hate_speech['new_token_sentence'].apply(FreqDist) #to check the frequency of each word
```

```python
frequency
```

```python
from wordcloud import WordCloud, STOPWORDS

wc = WordCloud(stopwords = set(list(STOPWORDS) + ['inside']), random_state = 100)
```

```python
fig, axes = plt.subplots(2, 2,squeeze=True, figsize=(20, 12))
axes = [ax for axes_row in axes for ax in axes_row]
for i, sub_topic_name in enumerate(['normal', 'offensive', 'hatespeech']):
  sub_topic = hate_speech[hate_speech[sub_topic_name] == 1]
  op = wc.generate(str(sub_topic['token_sentence']))
  ax = axes[i].imshow(op)
  ax2 = axes[i].set_title(sub_topic_name.upper(), fontsize=24)
  ax3 = axes[i].axis('off')
ax4 = plt.suptitle('Top words for particular category', fontsize=28)
```
